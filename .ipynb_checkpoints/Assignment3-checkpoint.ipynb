다음과 같이 0 보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.

1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오. )

```python
def addall(list):
    sum = 0
    for l in list:
        sum += l
    return sum
print addall([1])
def addall(list):
    return sum([l for l in list])
print addall([1,2,3,4,5,6,7,8,9])
```

2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~in 리터럴과 리스트 내보 방식으로 각각 제시하시오.)

```python
def addallodd(list):
    sum = 0
    for l in list:
        if l % 2 is 1:
            sum += l
    return sum
print addallodd([1])
def addallodd(list):
    return sum([l for l in list if l % 2 == 1])
print addallodd([1,2,3,4,5,6,7,8,9])
```

```python
import math
def isPrime(num):
    for i in range(2, int(math.sqrt(num)) + 1 ):
        if num % i == 0:
            return False
    return True
cnt = 0
num = 2
while cnt < 10001:
    if isPrime(num):
        cnt += 1
    num += 1
print num-1
```

```python
import urllib2
import string

def count_pure_word(url):
    source = urllib2.urlopen(url).read()

    for i in range(0, source.count('<!--')):
        if source.find("<!--") != -1:
            source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

    for i in range(0,source.count("<")):
        if source.find("<") != -1:
            source = source.replace(source[source.find("<"):source.find(">")+1:],"")

    words = source.split()

    my_str = " ".join(words)
    for s in string.punctuation:
        my_str = my_str.replace(s, "")

    new_words = my_str.split()
    new_dics = {}
    for word in new_words:
        if new_dics.has_key(word):
            new_dics[word] += 1
        else:
            new_dics[word] = 1
    return new_dics
```

```python
naver_dic = count_pure_word("http://www.naver.com")
daum_dic =  count_pure_word("http://www.daum.net") 
ny_dic =  count_pure_word("http://www.nytimes.com")
```
